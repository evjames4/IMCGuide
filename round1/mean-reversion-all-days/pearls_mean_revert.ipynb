```python
# Filtered to just Pearsl
df2 = pd.read_csv('../data/prices_round_1_day_-2.csv', index_col='timestamp', delimiter=';')
df1 = pd.read_csv('../data/prices_round_1_day_-1.csv', index_col='timestamp', delimiter=';')
df0 = pd.read_csv('../data/prices_round_1_day_0.csv', index_col='timestamp', delimiter=';')
df2 = df2.loc[df2['product']=='PEARLS']
df1 = df1.loc[df1['product']=='PEARLS']
df0 = df0.loc[df0['product']=='PEARLS']
df2 = df2[['mid_price']]
df1 = df1[['mid_price']]
df0 = df0[['mid_price']]
```
